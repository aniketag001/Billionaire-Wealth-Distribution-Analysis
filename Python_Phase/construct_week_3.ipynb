```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Billionaires Statistics Dataset.csv")

# Display the first few rows
df.head()
```

```python
print("Dataset Shape:", df.shape)
print("\nColumns:")
print(df.columns.tolist())
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
print("Dataset Shape:", df.shape)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['finalWorth'], bins=30, kde=True, color='goldenrod')
plt.title("Distribution of Billionaires' Final Net Worth")
plt.xlabel('Net Worth (USD)')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=20, kde=True, color='skyblue')
plt.title('Age Distribution of Billionaires')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(data=df, y='category', order=df['category'].value_counts().index, palette='viridis')
plt.title('Billionaires by Industry Category')
plt.xlabel('Number of Billionaires')
plt.ylabel('Category')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='selfMade', palette='pastel')
plt.title('Self-Made vs Inherited Billionaires')
plt.xlabel('Self-Made (True/False)')
plt.ylabel('Count')
plt.show()
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

# Compute the correlation matrix
corr_matrix = df[numeric_cols].corr()

plt.figure(figsize=(18, 16))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title("Full Correlation Heatmap of Numeric Features")
plt.show()
```

```python
avg_networth_by_country = df.groupby('country')['finalWorth'].mean().sort_values(ascending=False)

plt.figure(figsize=(24, 12))
sns.barplot(x=avg_networth_by_country.index, y=avg_networth_by_country.values, palette='magma')
plt.xticks(rotation=90)
plt.title("Average Net Worth by Country")
plt.xlabel("Country")
plt.ylabel("Average Net Worth (USD)")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='gdp_country', y='finalWorth', hue='category', palette='deep', alpha=0.7)
plt.title("GDP of Country vs. Billionaire Net Worth")
plt.xlabel("GDP of Country")
plt.ylabel("Final Net Worth (USD)")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
